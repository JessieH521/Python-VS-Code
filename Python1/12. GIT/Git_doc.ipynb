## VCS-operationer 
"版本控制系统操作"（Version Control System operations）
VCS（版本控制系统）是管理代码或文档变更的一种工具，它跟踪文件的每个修改，帮助开发者管理项目的不同版本，尤其在多人协作开发时，确保每个人都可以看到和合并变更。
常见的 VCS 操作包括：

Clone（克隆）：从远程仓库复制项目到本地计算机。

例子：git clone <repository-url>
Add（添加）：将文件的修改或新增文件添加到暂存区（staging area），以便之后提交。

例子：git add <filename>
Commit（提交）：将暂存区的修改保存到本地仓库，形成一个历史快照。

例子：git commit -m "Add new feature"
Push（推送）：将本地提交的更改推送到远程仓库，使其他人也可以访问这些更改。

例子：git push origin main
Pull（拉取）：从远程仓库获取最新的代码更新并合并到本地。

例子：git pull origin main
Branch（分支）：创建一个分支来独立开发，保证主分支不被影响，直到准备好合并为止。

例子：git checkout -b <branch-name>
Merge（合并）：将某个分支的修改合并到主分支或另一个分支。

例子：git merge <branch-name>
Revert（回滚）：回滚到之前的某个提交，取消某些更改。

例子：git revert <commit-hash>
Log（日志）：查看项目的提交历史，帮助开发者了解之前的更改。

例子：git log
VCS 操作允许团队协作开发，同时可以跟踪和管理代码的不同版本。常见的版本控制系统包括 Git、Subversion（SVN）、Mercurial 等，其中 Git 是最流行的。

## Vanliga VCS-operationer
• Initialisera: Sätt upp ett nytt repository. 
• Commit: Registrera ändringar som gjorts i filer. 
• Uppdatera/Pull: Få de senaste ändringarna från ett annat förråd eller gren. 
• Merge: Kombinera ändringar från olika grenar eller förråd. 
• Branch: Skapa en ny utvecklingslinje separat från den nuvarande eller huvudlinjen. 
• Checkout: Byt mellan olika versioner eller grenar. 
• Clone: Göra en fullständig kopia av ett repository.

## 常见的 VCS 操作
• 初始化：设置新的存储库。
• 提交：记录对文件所做的更改。
• 更新/拉取：从另一个存储库或分支获取最新更改。
• 合并：合并来自不同分支或存储库的更改。
• 分支：创建与当前或主线分开的新开发线。
• Checkout：在不同版本或分支之间切换。
• 克隆：制作存储库的完整副本。

main（主分支）
feature-branch（功能分支）
你在 feature-branch 上开发新功能，完成后希望将这个功能合并到 main 分支。那么你可以：

切换到 main 分支：git checkout main
合并功能分支：git merge feature-branch
合并后，main 分支会包含 feature-branch 的所有更改。

解决冲突（如果有）： 如果两个分支修改了相同的文件或相同的代码段，Git 可能会产生冲突。这时你需要手动解决冲突，然后提交合并结果。

### Vanliga Git-kommandon(命令)
• git init : Initialisera ett nytt Git-repository. 
• git clone [url] : Klona ett repository från en extern källa. 
• git add [file/directory] : Lägg till ändringar i staging-området. 
• git commit -m "[message] : Förbind dina ändringar. 
• git pull [remote] [branch] : Hämta och slå ihop ändringar från ett remote repository eller branch 
• git push [remote] [branch] : Skjut up dina ändringar till ett remote repository eller branch. 
• git branch : Lista, skapa eller ta bort branches. 
• git checkout [branch] : Byt till en annan branch.

## GitHub 网站：

1. 注册 GitHub 账号
前往 GitHub 官网。
点击右上角的 Sign up 按钮，创建一个新账号。
完成注册流程，并验证邮箱。
2. 创建新仓库 (Repository)
登录 GitHub 后，点击右上角的 + 按钮，选择 New repository。
填写仓库名、描述（可选），选择仓库公开（Public）或私有（Private）。
可以勾选 Initialize this repository with a README，这样仓库会自动生成一个README文件。
点击 Create repository 创建。
3. 上传代码
如果你已有代码在本地，想要上传到 GitHub：
打开终端，进入项目文件夹。
运行以下命令，将代码添加到 GitHub：
bash

git init                          # 初始化本地仓库
git add .                         # 添加所有文件到暂存区
git commit -m "Initial commit"    # 提交代码
git branch -M main                # 设置主分支为 main
git remote add origin https://github.com/你的用户名/你的仓库名.git  # 链接远程仓库， origin 是远程仓库别名
git push -u origin main           # 推送代码到 GitHub

如果有错误提示表明，你在尝试推送（git push）本地 main 分支到远程仓库时，远程仓库中的 main 分支有一些你本地没有的更新。Git 阻止了推送，以避免覆盖这些远程的更改。
可以先把远程仓库中的最新更改拉取到本地，然后再进行推送：
git pull origin main
git push origin main
如果pull la w的时候还提示错误可以：
重叠提交 (rebase)：将你的本地提交在远程更改的基础上重新应用，从而保持提交历史线性。适合保持干净的历史记录。
git config pull.rebase true
git pull origin main --rebase
git push origin main     ---> 上传到远程网站成功！

创建新分支：开发新功能或修复 bug 时，可以在新分支上进行工作：
bash

7. Pull Request (合并请求)
在新分支上完成更改后，可以通过 GitHub 提交 Pull Request，请求将该分支的更改合并到主分支。
在项目页面，点击 Pull Requests，选择 New Pull Request，创建新的合并请求。
经过审查后，可以点击 Merge 按钮，将代码合并。

GitHub 允许多个开发者在同一项目上协作。你可以：

查看贡献者和提交记录：通过 Insights 标签，了解项目的贡献者和开发历史。

在仓库中，可以通过创建 .yml 配置文件来定义工作流，自动执行任务。
10. 查看和管理代码
在 GitHub 网站上，你可以查看提交记录（Commits）、代码差异（Diffs），并通过 History 了解文件的变化。
你还可以直接在线编辑文件，进行简单的修改。

## 案例 GitHub
GIT i grupp
• 1. En skapar ett projekt med en fil hello.py
• 2. Invajta alla
• 3. Alla clonar
• Var och en ska:
- skapa en feature bransch
- lägga till en rad ”Hej från <aladdin>”
- commit + push
- efter OK -> Merge to main

1. 创建项目和文件
创建一个新项目目录：

mkdir mitt_projekt
cd mitt_projekt
初始化 Git 仓库：

git init
创建 hello.py 文件：
在项目目录中创建 hello.py 文件，内容可以是一个简单的问候语。

echo 'print("Hello from the project!")' > hello.py
提交初始文件：

git add hello.py
git commit -m "Initial commit with hello.py"
2. 邀请团队成员
确保你的代码托管在一个远程仓库（如 GitHub、GitLab 等）。

将本地仓库链接到远程仓库：
git remote add origin <repository-url>
推送到远程仓库：
git push -u origin main
共享仓库的链接，邀请团队成员加入。
3. 团队成员操作
每个团队成员按照以下步骤进行操作：

1. 克隆远程仓库
每个成员需要克隆远程仓库：

git clone <repository-url>
cd mitt_projekt
2. 创建特性分支
每个成员在本地仓库中创建一个特性分支：
git checkout -b feature_<你的名字>
3. 修改文件
在 hello.py 文件中添加一行代码。例如：
print("Hej från <Jie>")
（将 <aladdin> 替换为每个团队成员的名字）

4. 提交更改
添加更改：
git add hello.py
提交更改：
git commit -m "Added greeting from <aladdin>"
5. 推送更改到远程仓库
git push origin feature_<你的名字>
6. 合并到主分支
每个人完成推送后，可以选择一个人负责合并。合并步骤如下：

切换到主分支：
git checkout main
拉取最新的更改：
git pull origin main
git merge feature_<aladdin>  # 替换为实际的特性分支名称
推送合并后的主分支到远程仓库：
git push origin main
7. 处理合并冲突
如果在合并时出现冲突：

Git 会标记出冲突的地方，打开相关文件手动解决冲突。
解决冲突后，添加并提交：
git add hello.py  # 添加解决后的文件
git commit -m "Resolved merge conflict"

### 在 GitHub 中将分支合并到主分支（通常是 main 或 master）的步骤如下：

1. 确保您的分支已推送到 GitHub
在合并之前，您需要确保您想要合并的分支已经推送到 GitHub。您可以使用以下命令将分支推送到远程仓库：

bash
Kopiera kod
git push origin your-feature-branch
将 your-feature-branch 替换为您实际的分支名称。

2. 创建合并请求（Pull Request）
登录到 GitHub，然后导航到您项目的页面。
在页面的顶部，您会看到一个 Pull requests 选项卡，点击它。
点击 New pull request 按钮。
在 “Compare” 下拉框中，选择您想要合并的分支（例如 your-feature-branch）。
确保您要将其合并到的分支（例如 main）在另一个下拉框中正确选择。
GitHub 会显示出即将合并的更改。如果一切正常，点击 Create pull request 按钮。
输入合并请求的标题和描述，然后点击 Create pull request。
3. 合并请求审查
如果您的团队在代码审查过程中，您可以等待团队成员审核您的代码。审查者可以在合并请求中留下评论。
在没有问题的情况下，您或审查者可以合并该请求。
4. 合并分支
如果您是合并请求的创建者或者具有合并权限，您会看到一个 Merge pull request 按钮，点击它以将更改合并到主分支。
GitHub 可能会提示您选择合并方式：
Merge commit：将合并保留为单独的提交。
Squash and merge：将所有提交压缩成一个提交。
Rebase and merge：将提交重新排列并合并到主分支。
选择适合您团队工作流程的合并方式，然后确认合并。

5. 删除分支（可选）
合并后，您可以选择删除已合并的分支。通常，GitHub 会提供一个 Delete branch 按钮。这样可以保持仓库的整洁。

6. 更新本地仓库
完成合并后，别忘了更新您的本地 main 分支：
